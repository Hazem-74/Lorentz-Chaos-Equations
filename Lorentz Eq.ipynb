# __Lorentz Chaos equations__


The **Lorenz equations** are a system of three nonlinear ordinary differential equations that were originally derived by **Edward Lorenz** in 1963 as a simplified model for atmospheric convection. These equations are famous for being one of the first systems to exhibit **chaotic behavior**, meaning that small differences in initial conditions can lead to drastically different outcomesâ€”a phenomenon popularly known as the **butterfly effect**.

### Lorenz Equations (Standard Form):

$$
\begin{aligned}
\frac{dx}{dt} &= \sigma(y - x) \\
\frac{dy}{dt} &= x(\rho - z) - y \\
\frac{dz}{dt} &= xy - \beta z
\end{aligned}
$$

Where:

- $ x, y, z $: state variables  
  - $ x $: proportional to the intensity of convective motion
  - $ y $: related to temperature difference between ascending and descending fluid
  - $ z $: proportional to the distortion from vertical temperature profile
- $ \sigma $: Prandtl number (ratio of kinematic viscosity to thermal diffusivity)
- $ \rho $: Rayleigh number (dimensionless parameter controlling convection)
- $ \beta $: geometric factor


### Classic Parameter Values:

For chaotic behavior, Lorenz used these values:

- $ \sigma = 10 $
- $ \rho = 28 $
- $ \beta = \frac{8}{3} $

With these parameters, the system exhibits chaotic behavior for most initial conditions.

### Key Features of Lorenz System:

- **Nonlinearity**: The terms like $ xz $ and $ xy $ make the system nonlinear.
- **Dissipation**: Volume in phase space contracts over time (system is dissipative).
- **Strange Attractor**: Trajectories settle onto a complex fractal structure called the **Lorenz attractor**.
- **Sensitivity to Initial Conditions**: Small changes in starting points lead to widely divergent trajectories.


```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def lorenz(t, xyz, sigma, rho, beta):
    x, y, z = xyz
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parameters
sigma = 10
rho = 28
beta = 8/3

# Initial condition
xyz0 = [0, 1, 1.05]

# Time span and solution points
t_span = [0, 50]
t_eval = np.linspace(*t_span, 100000)

# Solve
sol = solve_ivp(lorenz, t_span, xyz0, args=(sigma, rho, beta), t_eval=t_eval)
```

```python
# Plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2], lw=0.5)
ax.set_title("Lorenz Attractor")
plt.show()
```
